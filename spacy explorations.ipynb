ok. so, we can look at the token in each of these to figure out profiles.

Write a routine to go through the nlp atributes of each one, and then through the tokens with each tweet. build up a list of frequent tokens.

```python
def getTokenCounts(tweets):
    counts={}
    for id,tweet in tweets.items():
        doc = tweet['nlp']
        for token in doc:
            if token.is_stop == False:
                if token.lemma_ not in counts:
                    counts[token.lemma_]=0
                counts[token.lemma_] = counts[token.lemma_] +1
    return counts
```

```python
smokeCounts=getTokenCounts(smoking)
```

```python
def getTopTokens(tokenCounts):
    ts=[]
    for t,count in tokenCounts.items():
        ts.append((t,count))
    ts.sort(key=lambda x: x[1],reverse=True)
    return ts
```

```python
topSmoke=getTopTokens(smokeCounts)
```
